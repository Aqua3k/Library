```python
"""
◆About
    multiset
    順序付きプライオリティキュー

◆各メソッドと計算量
    insert(value)
        値を追加 O(1)
    get_min()
        最小値を取得(削除はしない) O(log(N))
    erase(value)
        値を削除する O(log(N))
    is_exist(value)
        値が存在するか確認する O(1)

◆注意点
    最大値を扱うときはheapqのときと同様、各要素に-1をかけること
    削除の前にis_exist()で存在するか確かめること

◆その他
    参考: https://tsubo.hatenablog.jp/entry/2020/06/15/124657
"""
import heapq
from sys import exit
class HeapDict:
    def __init__(self):
        self.h=[]
        self.d=dict()
        self.length=0
    
    def __bool__(self):
        return self.length>=1

    def insert(self,x):
        heapq.heappush(self.h,x)
        self.length+=1
        if x not in self.d:
            self.d[x]=1
        else:
            self.d[x]+=1

    def erase(self,x):
        self.length-=1
        if x not in self.d or self.d[x]==0:
            print(x,"is not in HeapDict")
            exit()
        else:
            self.d[x]-=1

        while len(self.h)!=0:
            if self.d[self.h[0]]==0:
                heapq.heappop(self.h)
            else:
                break

    def is_exist(self,x):
        if x in self.d and self.d[x]!=0:
            return True
        else:
            return False

    def get_min(self):
        return self.h[0]

    def len(self):
        return self.length
```
